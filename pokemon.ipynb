```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("pokemon.csv")
```

```python
fig, axes = plt.subplots(1,2, figsize = (12,6)) # 1 row, 2 columns
axes[0].scatter(df["Attack"], df["Defense"], color = 'lavender', alpha = 0.5)
axes[0].set_xlabel("Attack")
axes[0].set_ylabel("Defense")
axes[0].set_title("Attack vs Defense for Pokemon")
axes[0].set_ylim(0, 200)
axes[0].set_xlim(0, 200)

sns.countplot(x= "Generation", data = df, ax = axes[1], palette="magma")
axes[1].set_xlabel("Generation")
axes[1].set_ylabel("Count")
axes[1].set_title("Generations of Pokemon")
axes[0].set_ylim(0, 200)

```
